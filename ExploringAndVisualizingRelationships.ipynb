```python
import matplotlib 
import pandas as pd

import matplotlib.pyplot as plt
```

```python
crude_oil_data = pd.read_csv('C:/Users/aanand/Desktop/Python-Data/crude_oil_data_processed.csv')
crude_oil_data.head(5)
```

```python
crude_oil_data['Date'] = pd.to_datetime(crude_oil_data['Date'])

crude_oil_data['Date'].head()
```

<h2> Simple line graph to show how US crude oil has varied across the years</h2>

```python
crude_oil_data.plot(x='Date', y='U.S. Crude Oil ', 
                    figsize=(12,8), color='blue')

plt.ylabel('Production')
plt.title('U.S Crude Oil Production')
```

```python
crude_oil_data.boxplot('U.S. Crude Oil ', 
                    figsize=(12,8))

plt.ylabel('Production')
plt.title('U.S Crude Oil Production')
```

```python
crude_oil_data[['California']].boxplot(figsize=(12,8))

plt.ylabel('Production in California')
plt.title('California Crude Oil Production')
```

<h3> The data points in the above boxplot are outliers </h3>

```python
crude_oil_data[['Alaska','California']].boxplot(figsize=(12,8))

plt.ylabel('Production in Alaska, California')
plt.title('Alaska, California Crude Oil Production')
```

<h3> It shows that the median production in Alaska and California are more or less same, there are more variation in the Crude oil production in Alaska. In addition, there 
are certain months in Alaska which are very high productivity months (positive outliers). And, with California, we have certain months which are very low productivity months
(negative outliers)</h3>

```python
crude_oil_data.boxplot(column=['U.S. Crude Oil '], by=['Year'],
                      figsize=(12,8))

plt.ylabel('Production')
```

<h2> Analyzing Crude oil production in the US on a yearly basis [groupby] function </h2>

```python
year_data = crude_oil_data.groupby('Year', as_index=False).sum()

year_data[['Year', 'U.S. Crude Oil ']]
```

```python
colors = ['C0', 'C1','C2','C3','C4','C5','C6','C7','C8','C9',]

year_data.plot.bar(x='Year', y = 'U.S. Crude Oil ',
                  figsize=(12,8), color = colors,
                  legend=False)

plt.ylabel('Production')
plt.title('U.s Crude Oil Production')
```

<h3> Findings for above graph: Seems like total oil production grew every year except in 2008 and 2018. Its because the dataset contains just 6 months worth of data, we dont
    have information on the entire year of production. Year 2015 and 2017 seems to be at the top in the production list</h3>

```python
crude_oil_data.columns
```

```python
'''Average crude oil data by states leaving the first and last 3 columns'''
mean_prod_data = crude_oil_data.mean()[1:-3]

mean_prod_data
```

```python
'''Sorting the result'''
mean_prod_data = mean_prod_data.sort_values(ascending=False)

mean_prod_data
```

```python
'''Creating a new dataframe to play with the columns above'''

mean_prod_df = pd.DataFrame(mean_prod_data).reset_index()
mean_prod_df.columns = ['State','Production']

mean_prod_df.head(10)
```

```python
plt.figure(figsize=(12,8))

plt.bar(mean_prod_df['State'], mean_prod_df['Production'],
       width=0.85)

plt.title('US Oil Mean Production June 2008 to June 2018')

plt.xticks(rotation = 90)
plt.xlabel('State')
plt.ylabel('Production')

plt.show()
```

<h2> Visualazing this oil production info across states in the form of probability density distribution</h2>

```python
'''KDE- Kernel Desnity Estimation which shows the probability distribution of oil production across states in the US'''
mean_prod_df['Production'].plot.kde(figsize=(12,8))

plt.title('US Oil Mean Production June 2008 to June 2018')
plt.xlabel('Oil Production')

plt.show()
```

```python
'''As our previous graphs has shown us that Texas is produces the max amount of crude oil, lets see its comparison with the total crude oil production'''

plt.figure(figsize=(12,8))

plt.scatter(crude_oil_data['Texas'], crude_oil_data['U.S. Crude Oil '], c='b')

plt.xlabel('US Production')
plt.ylabel('Texas Production')

plt.show()
```
