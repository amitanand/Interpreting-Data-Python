```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
automobile_data_processed = pd.read_csv('C:/Users/aanand/Desktop/Python-Data/automobile_data_processed.csv')
automobile_data_processed.head(5)
```

```python
'''How the price of a car is distributed based on the fuel-type - univariate data/relationship'''
plt.figure(figsize=(18,12))

fg = sns.FacetGrid(automobile_data_processed, col = 'fuel-type',
                  height = 7, aspect = 1)

fg.map(plt.hist, 'price', color='blue')

plt.show()
```

```python
'''How the price of a car is distributed based on the fuel-type - bivariate data/relationship'''
plt.figure(figsize=(18,12))

fg = sns.FacetGrid(automobile_data_processed, col = 'fuel-type',
                  height = 7, aspect = 1)

fg.map(sns.scatterplot, 'highway-mpg','price', color='blue', s= 100)
fg.add_legend()
plt.show()
```

```python
plt.figure(figsize=(18,12))

fg = sns.FacetGrid(automobile_data_processed, col = 'fuel-type',
                  hue='num-of-cylinders',height = 7, aspect = 1)

fg.map(sns.scatterplot, 'highway-mpg','price', s= 100)
fg.add_legend()
plt.show()
```

```python
tips_data = pd.read_csv('C:/Users/aanand/Desktop/Python-Data/tips.csv')

tips_data.sample(10)
```

```python
plt.figure(figsize=(18,12))

fg = sns.FacetGrid(tips_data, col = 'size',
                  height = 7, aspect = 1)

fg.map(plt.hist, 'total_bill', color='blue')

plt.show()
```

```python
group_size_values = np.sort(tips_data['size'].unique())

group_size_values
```

```python
fg = sns.FacetGrid(tips_data,
                  row='time',
                  height=6, aspect =1)

fg.map(sns.swarmplot,
      'size',
      'total_bill',
      order = group_size_values)

fg.add_legend()
```

```python
fg = sns.FacetGrid(tips_data,
                  col='time',
                  height=6, aspect =1)

fg.map(sns.swarmplot,
      'size',
      'total_bill',
      order = group_size_values)

fg.add_legend()
```
