```python
import seaborn as sns

sns.__version__
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
automobile_data = pd.read_csv('C:/Users/aanand/Desktop/Python-Data/Automobile_data.csv', na_values = '?')

automobile_data.head(5)
```

```python
automobile_data.shape
```

```python
'''isnull().sum() will give the column-vise sum of missing values'''
automobile_data.isnull().sum()
```

```python
automobile_data.dropna(inplace = True)

automobile_data.shape
```

```python
automobile_data.to_csv('C:/Users/aanand/Desktop/Python-Data/automobile_data_processed.csv', index = False)
```

```python
'''The easiest way to view the univariate distribution that exists in your data is to use seaborn distplot (distribution plot). Most of the vehicle's price falls in the 
range of 7000- 10,000 range'''

plt.figure(figsize=(12,8))

sns.distplot(automobile_data['price'],
            color = 'red')

plt.title('Automobile data')
plt.show()
```

```python
'''Making the buckets of 20'''
plt.figure(figsize=(12,8))

sns.distplot(automobile_data['price'],
            bins = 20, color = 'red')

plt.title('Automobile data')
plt.show()
```

```python
'''Showwing just KDE curve, not the histograms'''
plt.figure(figsize=(12,8))

sns.distplot(automobile_data['price'],
            hist = False, color = 'red')

plt.title('Automobile data')
plt.show()
```

```python
'''If you want to visualize actual data points or sample that exists in the data .. put rug = Ture and it overlays the rug plot that shows the individual data points
in the form of lines'''
plt.figure(figsize=(12,8))

sns.distplot(automobile_data['price'],
            hist = False, rug = True, color = 'red')

plt.title('Automobile data')
plt.show()
```

```python
'''Just the rugplot'''
plt.figure(figsize=(12,8))

sns.rugplot(automobile_data['price'],
            height = 0.5, color = 'red')

plt.title('Automobile data')
plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.kdeplot(automobile_data['price'],
            shade = True, color = 'red')

plt.title('Automobile data')
plt.show()
```

```python
'''Visualizing bivariate statistical relationships via scatterplot'''

plt.figure(figsize=(12,8))

sns.scatterplot(x='horsepower', y = 'price',
               data = automobile_data, s=120)

plt.title('Automobile Data')
plt.show()
```

<h2> We can see in the above scatterplot that the vehicles who has more horsepower has more price .. maybe because it needs more sophisticated engines </h2>

```python
'''COloring the datapoints based on the num of cylinders ... to do that, I will use hue parameter'''
plt.figure(figsize=(12,8))

sns.scatterplot(x='horsepower', y = 'price',
               hue= 'num-of-cylinders', data = automobile_data, s=120)

plt.title('Automobile Data')
plt.show()
```

<h2>The above scatterplot shows that the number of vehicles with 4 cylinders are maximum. It also shows that vehicles with higher horsepower have more cylinders and are
expensive </h2>

```python
'''Positive relationship between price and horsepower'''
plt.figure(figsize=(12,8))
sns.regplot(x='horsepower', y ='price',
           data = automobile_data)

plt.show()
```

```python
'''Negative relationship between price and highway-mpg'''
plt.figure(figsize=(12,8))
sns.regplot(x='highway-mpg', y ='price',
           data = automobile_data)

plt.show()
```

```python
'''Viewing univariate and bivariate relationships in the same graph'''

plt.figure(figsize = (12,8))

sns.jointplot(x='horsepower', y ='price',
             data = automobile_data)

plt.show()
```

```python
plt.figure(figsize = (12,8))

sns.jointplot(x='horsepower', y ='price',
             data = automobile_data, kind = 'reg')

plt.show()
```

```python
plt.figure(figsize = (12,8))

sns.jointplot(x='horsepower', y ='price',
             data = automobile_data, kind = 'kde')

plt.show()
```

```python
plt.figure(figsize = (12,8))

sns.jointplot(x='horsepower', y ='price',
             data = automobile_data, kind = 'hex')

plt.show()
```
